```python
from random import choice           #The choice function will be used to select a word from a list of words.

word_list = ['exchange','ghana','australia','cobweb', 'hacksaw','library','fireplace','memory','random','surgery',
            'gymnastics','embezzle','venezuela','decrypt','knapsack','schizophrenia','melancholy','elephant','discussion','rainbow']
dictionary = {'exchange':'giving and receiving','ghana':'Country in West Africa','australlia':'Continent','cobweb':'spider\'s web',
             'hacksaw':'for cutting metal','library':'collection of books','fireplace':'place in a room, that burns','memory':'something remembered from the past',
             'random':'unspecified','surgery':'branch of medicine','gymnastics':'Modern Sport', 'embezzle': 'misappropriate funds', 'venezuela':'Country in South America',
             'decrypt':'make a coded message intelligible', 'knapsack':'Type of bag', 'schizophrenia':'Mental disorder',
             'melancholy':'a feeling of sadness','elephant':'Animal','discussion':'process of talking','rainbow':'colour display'}

word_copy = word_list[:]            #This creates a copy of  the original list. I am creating a copy because I want the words to removed once they are chosen, to prevent repetition. 

def man(life):
    if life == 9:
        print("9 lives left")
        print("  --------  ")
    elif life == 8:
        print("8 lives left")
        print("  --------  ")
        print("     O      ")
    elif life == 7:
        print("7 lives left")
        print("  --------  ")
        print("     O      ")
        print("     |      ")
    elif life == 6:
        print("6 lives left")
        print("  --------  ")
        print("     O      ")
        print("     |      ")
        print("    /       ")
    elif life == 5:
        print("5 lives left")
        print("  --------  ")
        print("     O      ")
        print("     |      ")
        print("    / \     ")
    elif life == 4:
        print("4 lives left")
        print("  --------  ")
        print("   \ O      ")
        print("     |      ")
        print("    / \     ")
    elif life == 3:
        print("3 lives left")
        print("  --------  ")
        print("   \ O /    ")
        print("     |      ")
        print("    / \     ")
    elif life == 2:
        print("2 lives left")
        print("  --------  ")
        print("   \ O /|   ")
        print("     |      ")
        print("    / \     ")
    elif life == 1:
        print("1 life left")
        print("Please don't let him die")
        print("  --------  ")
        print("   \ O_|/   ")
        print("     |      ")
        print("    / \     ")



print('Welcome!!!')
print('Get Ready..........')

score = 0                        #initializing score at the beginning of the game to 0
life = 10                        # 10 lives are available for the game
x = True                         #This variable will be used to break the game loop when it changes to False

while x == True:
    
    word = choice(word_copy)     #A word is chosen from the list
    word_copy.remove(word)       #this will remove each word that is chosen from the list, to avoid repetition
    l = len(word)    
    lists = [ '-' for i in range(l)]     #creates a list of '-' with the word length, this will serve as the ----- which will be printed out to show the user the length of the word.
    print()
    print('New word ---------------')
    print()
    for i in lists:
        print(i,end='')                 #This will be print of dashes with length equal to the length of the secret word
    print()
    print(f'Hint - {dictionary[word]}') #Hint will also be printed.
    diction = {}         #This dictionary is useful incase a letter appears more than once in a word, it will keep track of the index   

    while lists != list(word):
        print()
        let = input('enter a letter you think:>> ')
        lets = let.lower()    #whatever the user inputs will be converted to lower case
        if lets not in word or len(lets) != 1 :   # the len(lets) != will also ensure only one letter is accepted at a time.
            print('wrong!')
            life -= 1
            if life == 0:
                  x = False
                  break 
            else:
                man(life)
                for i in lists:
                    print(i,end='')
            
                  
        else:
            if lets not in diction:
                index = word.find(lets)
                diction[lets] = index
                lists[index] = lets
                for i in lists:
                    print(i,end='')
            else:
                if lists.count(lets) != word.count(lets):
                    ind = diction[lets]
                    index = word.find(lets,ind + 1)
                    diction[lets] = index
                    lists[index] = lets
                    for i in lists:
                        print(i,end='')
                else:
                    print('wrong!')
                    life -= 1
                    if life == 0:
                        x = False
                        break 
                    else:
                        man(life)
                        for i in lists:
                            print(i,end='')
                     
    
#     for i in lists:
#         print(i,end='')
    if lists == list(word):
        score += 1
    if score == 10:
        x = False
        break
        
if score == 10:
    print('----------------------------------')
    print('Congratulations! You won the game!')
else:
    print("You loose")
    print("You let a kind man die")
    print("  --------  ")
    print("     O_|    ")
    print("    /|\      ")
    print("    / \     ")
    print(f'Your score: {score}')
```
